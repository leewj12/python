```python
# 델타를  이용한  2차  배열  탐색
# 2차 배열의 한 좌표에서 4방향의 인접 배열 요소를 탐색하는 방법
# ary[0…n-1][0…n-1]
# dx[] ← {0, 0, -1, 1}; // 상하좌우 
# dy[] ← {-1, 1, 0, 0};
# for x, y from 0 to n-1 
#     for i from 0 to 3 {
#         testX ← x + dx[i];
#         testY ← y + dy[i];
#         test(ary[testY][testX]);
#         }
d = [
    ['a','b','c'],
    ['d','e','f'],
    ['g','h','i']
]
gx = len(d) #행 (가로).
gy = len(d[0]) #열 (세로).
dx = [-1, 1, 0, 0] #상, 하, 좌, 우.
dy = [0, 0, -1, 1]
x, y = 1, 1
for i in range(len(dx)):
    nx = x + dx[i]
    ny = y + dy[i]
    if (0 <=  nx < gx) and (0 <= ny < gy): #list의 크기를 벗어나지 않게.
        print(d[nx][ny])
print(d)
```

```python
# 연습문제 1
# 1. 5x5 정방행렬 리스트 생성
# 2. random 값은 1-100 무작위로 채움(1,100 포함)
# 3. 상,하,좌,우 값이 존재하는 셀에 대해서만 아래 내용을 수행
#    - 현재 셀과 상하좌우 셀의 차를 구한다.
#    - 각 차의 절대값들을 구해서 더한 합계를 출력한다.
# 4. 출력은 아래와 같은 형식 으로 해주세요
#  1) 리스트 초기값을 화면에 출력
#  2) a[1][1] 이웃요소들 차의 절대값 합계 :  30
#      a[1][2] 이웃요소들 차의 절대값 합계 :  92
import random
l = [[random.randint(1, 100) for _ in range(5)] for _ in range(5)] # [(1~100)사이 랜덤값을 range(5)번 출력] * range(5)
print(l)
gx = len(l) #행 = 가로.
gy = len(l[0]) #열 = 세로.
dx = [-1, 1, 0, 0]  #상, 하, 좌, 우
dy = [0, 0, -1, 1]
for x in range(gx):  # 행 범위
    for y in range(gy):  # 열 범위
        for i in range(len(dx)):  
            nx = x + dx[i]  # 행 이동
            ny = y + dy[i]  # 열 이동
            if (0 <=  nx < gx) and (0 <= ny < gy):  # 범위 설정
                print('l[%d],[%d] 이웃요소들의 차의 절대값 합계: '%(x,y),abs(l[x][y] - l[nx][ny])) # 절대값 출력
```
