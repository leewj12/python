```python
from urllib import request
from bs4 import BeautifulSoup
# Open the URL and parse the content
target = request.urlopen('http://www.kma.go.kr/weather/forecast/mid-term-rss3.jsp?stnId=108')
soup = BeautifulSoup(target, 'html.parser')
# Open a file in write mode
with open('weather_0812.txt', 'w', encoding='utf-8') as file:
    for location in soup.select('location'):
        # Prepare the data to be written to the file
        city = '도시: ' + location.select_one('city').string + '\n'
        weather = '날씨: ' + location.select_one('wf').string + '\n'
        min_temp = '최저기온: ' + location.select_one('tmn').string + '\n'
        max_temp = '최고기온: ' + location.select_one('tmx').string + '\n'
        separator = '\n'
        # Write the data to the file
        file.write(city)
        file.write(weather)
        file.write(min_temp)
        file.write(max_temp)
        file.write(separator)
```

```python
from urllib import request
from bs4 import BeautifulSoup
from flask import Flask
app = Flask(__name__)
@app.route('/')
def hello():
# urlopen() 함수로 기상청의 전국 날씨를 읽는다.
response = request.urlopen('http://www.kma.go.kr/weather/forecast/mid-term-rss3.jsp?stnId=108')
soup = BeautifulSoup(response, 'html.parser')  # BeautifulSoup을 사용해 웹 페이지를 분석.
# Weather_0812.txt 파일로 저장하기.
with open('Weather_0812.txt', 'w', encoding='UTF-8') as file:
    output = ''
    for location in soup.select('location'):  # location 태그를 찾는다.
        # 내부의 city, wf, tmn, tmx 태그를 변수로 만들어 놓기.
        city = location.select_one('city').string
        weather = location.select_one('wf').string
        tmn = location.select_one('tmn').string
        tmx = location.select_one('tmx').string
        # 1. 화면에 출력하기.
        # print(f'도시: {city}')
        # print(f'날씨: {weather}')
        # print(f'최저기온: {tmn}')
        # print(f'최고기온: {tmx}')
        # print()
        output += '<h3>{}<h3>'.format(location.select_one('city').string)
        output += '날씨: {}</br>'.format(location.select_one('wf').string)
        output += '최저/최고 기온:{}/{}\'
            .format(\
                location.select_one('tmn').string,\
                location.select_one('tmx').string\
            )
            output += '<hr/>'
        return output
        # 2. 파일로 쓰고 저장하기.
        file.write(f'도시: {city}\n')
        file.write(f'날씨: {weather}\n')
        file.write(f'최저기온: {tmn}\n')
        file.write(f'최고기온: {tmx}\n')
        file.write('\n')
```
