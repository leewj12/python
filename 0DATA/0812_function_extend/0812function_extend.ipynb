```python
def sum_all(start=0, end=100, step=1):
    output = 0
    for i in range(start, end + 1, step):
        output += i
    return output
print(sum_all())
print('A.', sum_all(0,100,10))
print('B.', sum_all(end=100))
print('C.', sum_all(end=100,step=2))
```

```python
def mul(*values):
    x = 1
    for i in values:
        x *= i
    return x

print(mul(5,7,9,10))
print(mul(8,3,5,7,9,10))
```

```python
def F(n):
    output = 1
    for i in range(1,n+1):
        output *= i
    return output
print('1!:', F(1))
print('2!:', F(2))
print('3!:', F(3))
print('4!:', F(4))
print('5!:', F(5))
print('12!: ', F(12))
```

```python
def F(n):
    if n ==0:
        return 1
    else:
        return n * F(n-1)
        
print('1!:', F(1))
print('2!:', F(2))
print('3!:', F(3))
print('4!:', F(4))
print('5!:', F(5))
print('12!: ', F(12))
```

```python
counter = 0
def fibonacci(n):
    print('fibonacci({})를 구합니다.'.format(n))
    global counter
    counter += 1
    if n == 1 or n == 2:
        return 1
    else: 
        return fibonacci(n-1) + fibonacci(n-2)
fibonacci(10)
print('---')
print('fibonacci(10)) 계산에 활용된 덧셈 횟수는 {}번 입니다.'.format(counter))
```

```python
d = {
    1: 1,
    2: 1
}
def f(n):
    if n in d:
        return d[n]
    else:
        output = f(n-1)+ f(n-2)
        d[n] = output
        return output

print('fibonacci(10):',f(10))
print('fibonacci(20):',f(20))
print('fibonacci(30):',f(30))
print('fibonacci(40):',f(40))
print('fibonacci(50):',f(50))
```

```python
import random
han = list('가나다라마바사아자차카타파하')
with open('info.txt','w') as file:
    for i in range(1000):
        name = random.choice(han) + random.choice(han)
        weight = random.randrange(40,100)
        height = random.randrange(140,200)
        file.write('{},{},{}\n'.format(name,weight,height))
```

```python
with open('info.txt', 'r') as file:
    for line in file:
        (name, weight, height) = line.strip().split(',')
        if (not name) or (not weight) or (not height):
            continue
        bmi = int(weight)/((int(height)/100)**2)
        result = ''
        if 25 <= bmi:
            result = '과체중'
        elif 18.5 <= bmi:
            result = '정상체중'
        else:
            result = '저체중'
        print('\n'.join([
            '이름:{}',
            '몸무게:{}',
            '키:{}',
            'BMI:{}',
            '결과:{}'
        ]).format(name,weight,height,bmi,result))
        print()
```
