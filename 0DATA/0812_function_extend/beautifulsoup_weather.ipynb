```python
from urllib import request
from bs4 import BeautifulSoup
# Open the URL and parse the content
target = request.urlopen('http://www.kma.go.kr/weather/forecast/mid-term-rss3.jsp?stnId=108')
soup = BeautifulSoup(target, 'html.parser')
# Open a file in write mode
with open('weather_0812.txt', 'w', encoding='utf-8') as file:
    for location in soup.select('location'):
        # Prepare the data to be written to the file
        city = ': ' + location.select_one('city').string + '\n'
        weather = ': ' + location.select_one('wf').string + '\n'
        min_temp = ': ' + location.select_one('tmn').string + '\n'
        max_temp = ': ' + location.select_one('tmx').string + '\n'
        separator = '\n'
        # Write the data to the file
        file.write(city)
        file.write(weather)
        file.write(min_temp)
        file.write(max_temp)
        file.write(separator)
```
