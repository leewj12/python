```python
import requests as req
from bs4 import BeautifulSoup as bs
import pandas as pd
import pyautogui as py

p = 1 #int(input('조회 할 마지막 페이지 번호 입력:'))
data = [] # 결과값 저장 리스트
for i in range(0,p+1):
    res = req.get(f'https://finance.naver.com/sise/sise_market_sum.naver?sosok=0&page={i}')
    html = res.text
    soup = bs(html,'lxml') # html 문서 번역된 소스
    
    stocks= soup.select('#contentarea > div.box_type_l > table.type_2 > tbody > tr')
    #contentarea > div.box_type_l > table.type_2 > tbody > tr:nth-child(2) > td:nth-child(2) > a
    #contentarea > div.box_type_l > table.type_2 > tbody > tr:nth-child(3) > td:nth-child(2) > a

    for stock in stocks:
        if len(stock) <=1 :#tr 공백에는 글자가 없기에 글자수가 0임 
            continue 

        title = stock.select_one('td:nth-child(2)') # 종목 == (td.tltle)
        price = stock.select_one('td:nth-child(3)') # 현재가격

        change = stock.select_one('td:nth-child(4)') # 전일대비
        ra = stock.select_one('td:nth-child(5)') # 등락률
        
        if title and price and change and ra:
            data.append({'종목명':title.text.strip(),'현재가':price.text.strip(),'전일비':change.text.strip(),'등락률':ra.text.strip()})
            
# df = pd.DataFrame(data) # pd를 이용해 행렬구조의 데이터프레임으로 저장
# df.set_index('종목명',inplace=True)
# df.to_csv('네이버주식시세_utf-8.csv',encoding='utf-8')
# df.to_csv('네이버주식시세_utf-8-sig.csv',encoding='utf-8-sig')
# df.to_csv('네이버주식시세_euc-kr.csv',encoding='euc-kr')
```
