```python
import requests as req
from bs4 import BeautifulSoup as bs
import pandas as pd
from selenium import webdriver  # 크롬드라이버 포함
from webdriver_manager.chrome import ChromeDriverManager # 크롬드라이버 자동업데이트

from selenium.webdriver.common.by import By # css selector를 활용해서 html 요소 가져오는 모듈
from selenium.webdriver.chrome.service import Service # chrome 드라이버 최신버전 설치

serv = Service(executable_pass = ChromeDriverManager().install())
driver = webdriver.Chrome(service = serv) # 크롬창을 만듦

driver.get('https://www.melon.com/chart/hot100/index.htm')  # 드라이버를 통해 url의 웹 페이지를 오픈

# for like in like_counts:
#     print(like.text) # 확인용.

head = {"user-agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/127.0.0.0 Safari/537.36"}
res = req.get("https://www.melon.com/chart/hot100/index.htm", headers=head)
html = res.text
soup = bs(html, 'lxml')

#lst50 > td:nth-child(2) > div > span.rank
ranks = soup.select("tbody span.rank")
#lst50 > td:nth-child(6) > div > div > div.ellipsis.rank01 > span > a
titles = soup.select(".ellipsis.rank01 a")
#lst50 > td:nth-child(6) > div > div > div.ellipsis.rank02 > span > a
singers = soup.select(".ellipsis.rank02 span.checkEllipsis")
#lst50 > td:nth-child(7) > div > div > div > a
albums = soup.select(".ellipsis.rank03 a")
#좋아요 갯수들 가져오기(동적 크롤링)
like_counts = driver.find_elements(By.CSS_SELECTOR, "tbody button span.cnt")
data= []

for i in range(len(ranks)):
    rank = ranks[i].text.strip()
    title = titles[i].text.strip()
    singer = singers[i].text.strip()
    album = albums[i].text.strip()
    like_count = like_counts[i].text.strip()
    data.append({"순위":rank,"제목":title, "가수":singer, "앨범":album, "좋아요":like_count})
    print(f'{rank}위 제목:{title} 가수:{singer} 앨범:{album} 좋아요:{like_count}')  

df = pd.DataFrame(data)
df.set_index("순위", inplace=True)
df.to_csv("멜론차트HOT100_utf-8.csv", encoding="utf-8")
df.to_csv("멜론차트HOT100_euc-kr.csv", encoding="euc-kr",errors='ignore')
df.to_csv("멜론차트HOT100_utf-8-sig.csv", encoding="utf-8-sig")
print("멜론차트HOT100 파일 생성 완료")
```
