```python
from selenium import webdriver  # webdriver = 크롬창 역할
from webdriver_manager.chrome import ChromeDriverManager # 크롬드라이버 자동업데이트
from selenium.webdriver.chrome.service import Service # 크롬 드라이버 최신버전으로 설치
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.keys import Keys # 컴퓨터용 키보드 라이브러리
from selenium.webdriver.common.by import By #요소 선택자를 구분해주는 라이브러
import pandas as pd
from bs4 import BeautifulSoup as bs
import time

a = input('검색어 입력')

# 크롬 드라이브 설치
serv = Service(executable_path = ChromeDriverManager().install())
#크롬창 생성
driver = webdriver.Chrome(service=serv)
url = f'https://www.youtube.com/results?search_query={a}'
driver.get(url)
time.sleep(2)
body = driver.find_element(By.TAG_NAME,'body')
try:
    driver.find_element(By.CSS_SELECTOR, '#scroll-container #chips yt-chip-cloud-chip-renderer:nth-child(6) yt-formatted-string').click()
    time.sleep(2)
except:
    pass
for i in range(20):
    body.send_keys(Keys.PAGE_DOWN)
    time.sleep(0.3)
html = driver.page_source
soup = bs(html,'lxml')
titles = soup.select('#video-title > yt-formatted-string')
views = soup.select('#metadata-line > span:nth-child(3)')
title_list =[]
view_list = []

for j in range(len(titles)):
    title_list.append(titles[j].text.strip())
    view_list.append(views[j].text.strip())
    print(title_list[j],view_list[j])

dic ={'제목':title_list,'조회수':view_list}
df = pd.DataFrame(dic)
df.set_index("제목", inplace=True)
df.to_csv("유튜브_utf-8.csv", encoding="utf-8")
df.to_csv("유튜브_euc-kr.csv", encoding="euc-kr",errors='ignore')
df.to_csv("유튜브_utf-8-sig.csv", encoding="utf-8-sig")
```
