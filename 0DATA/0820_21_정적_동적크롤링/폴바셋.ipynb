```python
from selenium import webdriver  # webdriver = 크롬창 역할
from selenium.webdriver.common.keys import Keys # 컴퓨터용 키보드 라이브러리
from selenium.webdriver.common.by import By #요소 선택자를 구분해주는 라이브러
from selenium.webdriver.chrome.service import Service # 크롬 드라이버 최신버전으로 설치
from selenium.webdriver.chrome.options import Options
from webdriver_manager.chrome import ChromeDriverManager # 크롬드라이버 자동업데이트
import pandas as pd
from bs4 import BeautifulSoup as bs
import time
# 크롬 드라이버 설치
serv = Service(executable_path = ChromeDriverManager().install())
# 크롬 창 생성
driver = webdriver.Chrome(service = serv)

t = 1
table =[]
url ='https://www.baristapaulbassett.co.kr/store/Store.pb'
search = driver.get(url)
time.sleep(t)

search = driver.find_element(By.LINK_TEXT, "지역").click()
time.sleep(t)

locals = driver.find_elements(By.CSS_SELECTOR,'#areaList > li')

for i in range(1, len(locals)+1):
    local = driver.find_element(By.CSS_SELECTOR,f'#areaList > li:nth-child({i})').text
    search = driver.find_element(By.CSS_SELECTOR,f'#areaList > li:nth-child({i})').click()
    time.sleep(t)
    driver.find_element(By.CSS_SELECTOR,'#countyList > li').click()
    time.sleep(t)
    html = driver.page_source
    soup = bs(html,'lxml')
    datas = soup.select('#shopList li')
    cnt = 0
    for data in datas:
        try:
            store = data.select_one('a strong').text.strip()
            add = data.select_one('address').text.strip()
            num = data.select_one('dl dd').text.strip()
            cnt += 1
            print(f"{local} {cnt:2} : {store} {add} {num}")
            table.append({'지역':local,
                        '지역별 매장':cnt,
                        '매장명' : store,
                         '주소' : add,
                          '전화번호':num
                         })
        except:
            pass
    driver.find_element(By.LINK_TEXT, "지역").click()

df = pd.DataFrame(table)
df.set_index('매장명', inplace=True)
df.to_csv("폴바셋_utf-8.csv", encoding="utf-8")
df.to_csv("폴바셋_utf-8-sig.csv", encoding="utf-8-sig")
df.to_csv("폴바셋_euc-kr.csv", encoding="euc-kr",errors = 'ignore')
```
