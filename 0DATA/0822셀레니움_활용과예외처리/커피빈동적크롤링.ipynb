```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from webdriver_manager.chrome import ChromeDriverManager
from bs4 import BeautifulSoup as bs
import time

serv = Service(executable_path=ChromeDriverManager().install())
driver = webdriver.Chrome(service=serv)

url = "https://www.coffeebeankorea.com/store/store.asp"
driver.get(url) #커피핀 사이트 창 띄움

start = driver.window_handles

print(driver.window_handles) # 커피빈 창 핸들값 1개만 출력됨

main_window = start[0]
print(main_window)
time.sleep(5)

#새 창 열기
# driver.execute_script('window.open("https://www.google.com");')
driver.execute_script('storePop2(31)') # 커피빈 상세 페이지
time.sleep(2)
end = driver.window_handles
print(end)
# ['95363F0799EBE8A07660221DFB2EE744', '49D5FF6FB06C76F2FB16836B9E52F775']

try:
    #새로 열린 창으로 전환
    for handle in end:
        if handle != main_window:
            driver.switch_to.window(handle)
            print(handle,'구글창')
            time.sleep(5)
            break
    
    # 새 창에서 동작 수행
    
    # 새 창 닫기
    driver.close()
    time.sleep(1)
    # 메인 창으로 전환
    driver.switch_to.window(main_window)
    time.sleep(5)
    print(driver.current_window_handle,'메인 커피빈 창')
    time.sleep(5)
    driver.quit()
except:
    pass
```
