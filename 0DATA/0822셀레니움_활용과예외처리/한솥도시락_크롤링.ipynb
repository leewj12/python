```python
from selenium import webdriver  # 셀레니움 모듈을 활용해 웹크롤링 기능 가져옴
from webdriver_manager.chrome import ChromeDriverManager # 크롬브라우저 설치를 해주는 드라이버
from selenium.webdriver.chrome.service import Service # 최신 버전의 크롬 드라이버 설치 지원

from selenium.webdriver.common.by import By # html 관련 요소들을 선택할 수 있는 모듈
from selenium.webdriver.chrome.options import Options
# from selenium.webdriver.common.keys import Keys 

from bs4 import BeautifulSoup as bs # html코드 해석(번역)
import time # 딜레이 시간을 설정
import pandas as pd # 행렬구조의 2차원 테이블 형태로 데이터 저장

# ChromeDriverManager를 이용해서 크롬드라이브를 설치
serv = Service(executable_path = ChromeDriverManager().install())
# service를 이용해서 최신 버전의 크롬브라우저를 만든 후 driver 변수에 할당
driver = webdriver.Chrome(service = serv)

driver.get('https://www.hsd.co.kr/menu/menu_list')
time.sleep(2)

try:
    for i in range(5):
        btn = driver.find_element(By.CSS_SELECTOR, "a.c_05")
        btn.click()
        time.sleep(1)
except:
    print('마지막 페이지입니다')

# title = driver.find_elements(By.CSS_SELECTOR,'.h.fz_03') #menuList_560 > div > div.item-text > h4
title = driver.find_elements(By.CSS_SELECTOR,' div.item-text > h4') #menuList_560 > div > div.item-text > div > strong
# price = driver.find_elements(By.CSS_SELECTOR,'.item-price') #menuList_559 > div > div.item-text > h4 
price = driver.find_elements(By.CSS_SELECTOR,'div.item-text > div > strong') #menuList_559 > div > div.item-text > div > strong

# print(len(title))

title_list = []
price_list = []

print('-----------------------------------------------------------------')
print('메뉴:','                                                      가격')
print('-----------------------------------------------------------------')
for i in range(len(title)):
    title_list.append(title[i].text.replace('[8월 할인메뉴]','').strip())
    price_list.append(price[i].text.strip())
    print(f"{title_list[i]:40}\t{price_list[i]:>10}")
    # print('%40s %10s'%(title_list[i],price_list[i]))

dic = {'메뉴명':title_list,'가격정보':price_list}
hs = pd.DataFrame(dic)
hs.set_index('메뉴명',inplace=True)
hs.to_csv('한솥메뉴_euc-kr.csv',encoding='euc-kr')
hs.to_csv('한솥메뉴_utf-8.csv',encoding='utf-8')
hs.to_csv('한솥메뉴_utf-8-sig.csv',encoding='utf-8-sig')
```

```python
from selenium import webdriver  # 셀레니움 모듈을 활용해 웹크롤링 기능 가져옴
from webdriver_manager.chrome import ChromeDriverManager # 크롬브라우저 설치를 해주는 드라이버
from selenium.webdriver.chrome.service import Service # 최신 버전의 크롬 드라이버 설치 지원

from selenium.webdriver.common.by import By # html 관련 요소들을 선택할 수 있는 모듈
from selenium.webdriver.chrome.options import Options
# from selenium.webdriver.common.keys import Keys 

from bs4 import BeautifulSoup as bs # html코드 해석(번역)
import time # 딜레이 시간을 설정
import pandas as pd # 행렬구조의 2차원 테이블 형태로 데이터 저장

import os # 폴더(디렉토리) 생성
from urllib.request import urlretrieve

# ChromeDriverManager를 이용해서 크롬드라이브를 설치
serv = Service(executable_path = ChromeDriverManager().install())
# service를 이용해서 최신 버전의 크롬브라우저를 만든 후 driver 변수에 할당
driver = webdriver.Chrome(service = serv)

driver.get('https://www.hsd.co.kr/menu/menu_list')
time.sleep(1)

try:
    for i in range(7):
        btn = driver.find_element(By.CSS_SELECTOR, "a.c_05")
        btn.click()
        time.sleep(1)
except:
    print('마지막 페이지입니다')

# title = driver.find_elements(By.CSS_SELECTOR,'.h.fz_03') #menuList_560 > div > div.item-text > h4
title = driver.find_elements(By.CSS_SELECTOR,' div.item-text > h4') #menuList_560 > div > div.item-text > div > strong
# price = driver.find_elements(By.CSS_SELECTOR,'.item-price') #menuList_559 > div > div.item-text > h4 
price = driver.find_elements(By.CSS_SELECTOR,'div.item-text > div > strong') #menuList_559 > div > div.item-text > div > strong

# 이미지를 저장할 폴더(디렉토리) 생성
if not os.path.isdir('c:/hansot'):
    os.mkdir('c:/hansot')

title_list = []
price_list = []
src_list = []

#menuList_560 > div > div.item-img > img
img = driver.find_elements(By.CSS_SELECTOR,'div.item-img > img')

print('-----------------------------------------------------------------')
print('메뉴:','                                                      가격')
print('-----------------------------------------------------------------')
for i in range(len(title)):
    title_list.append(title[i].text.replace('[8월 할인메뉴]','').strip())
    price_list.append(price[i].text.strip())
    src_list.append(img[i].get_attribute('src'))
    print(f"{title_list[i]:40}\t{price_list[i]:>10}")
    # print('%40s %10s'%(title_list[i],price_list[i]))

dic = {'메뉴명':title_list,'가격정보':price_list}

hs = pd.DataFrame(dic)
hs.set_index('메뉴명',inplace=True)

hs.to_csv('한솥메뉴_euc-kr.csv',encoding='euc-kr')
hs.to_csv('한솥메뉴_utf-8.csv',encoding='utf-8')
hs.to_csv('한솥메뉴_utf-8-sig.csv',encoding='utf-8-sig')

for j in range(len(src_list)):
    urlretrieve(src_list[j],'c:/hansot/'+title_list[j]+'.jpg') # urlretieve(이미지주소, 저장할 경로 및 파일명)
    
driver.quit()
```
