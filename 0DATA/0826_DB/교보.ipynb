```python
from selenium import webdriver  # 셀레니움 모듈을 활용해 웹크롤링 기능 가져옴
from webdriver_manager.chrome import ChromeDriverManager # 크롬브라우저 설치를 해주는 드라이버
from selenium.webdriver.chrome.service import Service # 최신 버전의 크롬 드라이버 설치 지원
from selenium.webdriver.common.by import By # html 관련 요소들을 선택할 수 있는 모듈
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.keys import Keys
from bs4 import BeautifulSoup as bs # html코드 해석(번역)
import time # 딜레이 시간을 설정
import pandas as pd # 행렬구조의 2차원 테이블 형태로 데이터 저장

# import os # 폴더(디렉토리) 생성
# from urllib.request import urlretrieve

# ChromeDriverManager를 이용해서 크롬드라이브를 설치
serv = Service(executable_path = ChromeDriverManager().install())
# service를 이용해서 최신 버전의 크롬브라우저를 만든 후 driver 변수에 할당
driver = webdriver.Chrome(service = serv)

book_l = [] # 책이름 리스트
author_l = [] # 저자 리스트
pub_l = [] # 출판사 리스트
date_l = [] # 출판일 리스트
price_l = [] # 가격 리스트
sale_l = [] # 할인가 리스트
score_l = [] # 평점 리스트

for a in range(1,11): # 페이지 범위 설정 1~ 10
    driver.get(f'https://store.kyobobook.co.kr/bestseller/total/weekly?page={a}') # 교보문고 주소창 
    time.sleep(2)
    
    book = driver.find_elements(By.CSS_SELECTOR, '.prod_link.font-text-xl') # 책이름 class 값으로 가져옴
    # prod_link font-weight-medium line-clamp-2 text-black hover:underline   font-text-xl mt-2
    
    author = driver.find_elements(By.CSS_SELECTOR,'.line-clamp-2.font-text-l.mt-1') # 저자, 출판사, 출판일 가져옴
    # line-clamp-2 flex overflow-hidden whitespace-normal break-all text-gray-800 font-text-l mt-1
    
    price = driver.find_elements(By.CSS_SELECTOR,'.flex.flex-col.mt-3') # 가격, 할인가 가져옴
    # .flex.flex-col.mt-3
    
    score = driver.find_elements(By.CSS_SELECTOR,'span.font-weight-bold.text-black') # 평점 가져옴
    # font-weight-bold text-black
    
    for i in range(len(book)):
        book_l.append(book[i].text.strip()) # 책이름 리스트에 추가
        print(book_l[i+((a-1)*20)])
        author_l.append(author[i].text.strip().split('·')[0].strip()) # 저자 리스트에 추가
        print(author_l[i+((a-1)*20)])
        pub_l.append(author[i].text.strip().split('·')[1].strip()) # 출판사 리스트에 추가
        print(pub_l[i+((a-1)*20)])
        date_l.append(author[i].text.replace('.','-').split('·')[2].strip()) # 출판일 리스트에 추가
        print(date_l[i+((a-1)*20)])
        price_l.append(price[i].text.replace(',','').replace('원','').split('\n')[2].strip()) # 가격 리스트에 추가
        print(price_l[i+((a-1)*20)])
        sale_l.append(price[i].text.replace(',','').replace('원','').split('\n')[1].strip()) # 할인가 리스트에 추가
        print(sale_l[i+((a-1)*20)])
        score_l.append(score[i].text.strip()) # 평점 리스트에 추가
        print(score_l[i+((a-1)*20)])
        print('---------------')
    
driver.close() # 창 닫기

dic = {'책이름':book_l,'저자':author_l,'출판사':pub_l,'출판일':date_l,'가격':price_l,'할인가':sale_l,'평점':score_l} # 리스트 딕셔너리화
df = pd.DataFrame(dic) # 데이터프레임 저장

df.set_index('책이름',inplace = True) # 책이름을 인덱스로 설정
df.to_csv('주간교보_euc-kr.csv',encoding = 'euc-kr') # euc-kr로 저장
df.to_csv('주간교보_utf-8.csv',encoding = 'utf-8') # utf-8로 저장
df.to_csv('주간교보_utf-8-sig.csv',encoding = 'utf-8-sig') # utf-8-sig로 저장
```
