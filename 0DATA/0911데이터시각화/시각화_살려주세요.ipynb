```python
#기본 사용 - fill_between() / 그래프 영역 채우기

x=[1,2,3,4]
y=[2,3,5,10]

plt.plot(x,y)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.fill_between(x[1:3], y[1:3], alpha=1)
#(x[1], y[1], (x[2], y[2]), (x[1], 0), (x[2], 0)을 잇는 영역이 채워짐

plt.show()
```

```python
x=[1,2,3,4]
y=[2,3,5,10]

plt.plot(x,y)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.fill_betweenx(y[2:4], x[2:4], alpha=0.5)
#(x[2], y[2], (x[3], y[3]), (0, y[2]), (0, y[3])을 잇는 영역이 채워짐

plt.show()
```

```python
#두 그래프 사이 영역 채우기

x = [1,2,3,4]
y1 = [2,3,5,10]
y2 = [1,2,4,8]

plt.plot(x,y1)
plt.plot(x,y2)
plt.xlabel('x')
plt.ylabel('y')
plt.fill_between(x[0:3], y1[0:3], y2[0:3], color='g', alpha=0.5)
# (x[0], y1[0]), (x[2], y1[2]), (x[2], y[2]), (x[2], y2[0]) 사이 영역 채움
# x[0]에서 x[2]까지, y1과 y2 사이의 영역을 채움
plt.show()
```

```python
# 수평선/수직선 표시

x = np.arange(0, 4, 0.5)

plt.plot(x, x + 1, 'bo')
plt.plot(x, x**2 - 4, 'g--')
plt.plot(x, -2*x + 3, 'r:')

plt.axhline(4.0, 0.1, 0.9, color='lightgray', linestyle='--', linewidth=4)
# axhline(y값, xmin, xmax) /xmin, xmax는 0에서 1사이 값. 0은 왼쪽 끝, 1은 오른쪽 끝
plt.hlines(-0.62, 1.0, 2.5, color='gray', linestyle='solid', linewidth=3)
# hlines(y, xmin, xmax) / 점 (xmin, y)에서 점 (xmax, y)를 따라 수평선을 표시

plt.show()
```

```python
# 수직선 그리기

x = np.arange(0, 4, 0.5)

plt.plot(x, x + 1, 'bo')
plt.plot(x, x**2 - 4, 'g--')
plt.plot(x, -2*x + 3, 'r:')

plt.axvline(1.0, 0.2, 0.8, color='lightgray', linestyle='--', linewidth=2)
# axvline(x 값, ymin, ymax) 
plt.vlines(1.8, -3.0, 2.0, color='gray', linestyle='solid', linewidth=3)
# vlines(x, ymin, ymax) / 점 (x, ymin)에서 점 (x, ymax)를 따라 수평선을 표시
plt.show()
```

```python
# 시작 각도와 방향 설정

ratio = [34, 32, 16, 18]
labels = ['Apple', 'Banana', 'Melon', 'Grapes']
explode = [0, 0, 0, 0.20]

plt.pie(ratio, labels=labels, autopct='%.1f%%', startangle=260, counterclock=False, explode=explode)
plt.show()
```

```python
# 중심에서 벗어나는 정도 설정 / 색상 지정

ratio = [34, 32, 16, 18]
labels = ['Apple', 'Banana', 'Melon', 'Grapes']
explode = [0, 0.10, 0, 0.10]
# explode는 부채꼴이 파이 차트의 중심에서 벗어나는 정도를 설정
# Banana’와 ‘Grapes’ 영역에 대해서 반지름의 10% 만큼 벗어나도록 설정
colors = ['silver', 'gold', 'lightblue', 'lightgray']

plt.pie(ratio, labels=labels, autopct='%.1f%%', startangle=260, counterclock=False, explode=explode, shadow=True, colors=colors)
#shadow를 True로 설정하면, 파이 차트에 그림자가 표시
plt.show()
```

```python
# 부채꼴 스타일 지정

ratio = [34, 32, 16, 18]
labels = ['Apple', 'Banana', 'Melon', 'Grapes']
colors = ['#ff9999', '#ffc000', '#8fd9b6', '#d395d0']
wedgeprops={'width': 0.5, 'edgecolor': 'w', 'linewidth': 2}
#‘wedgeprops 딕셔너리의 ‘width’, ‘edgecolor’, ‘linewidth’ 키를 이용해서 각각 부채꼴 영역의 너비 (반지름에 대한 비율), 테두리의 색상, 테두리 선의 너비를 설정

plt.pie(ratio, labels=labels, autopct='%.1f%%', startangle=260, counterclock=False, colors=colors, wedgeprops=wedgeprops)

plt.show()
```

```python
x1 = np.linspace(0.0, 5.0)
x2 = np.linspace(0.0, 2.0)

y1 = np.cos(2 * np.pi * x1) * np.exp(-x1)
y2 = np.cos(2 * np.pi * x2)

#subplot 위치들만 확인
plt.subplot(2, 1, 2)                # nrows=2, ncols=1, index=1
plt.plot(x1, y1, 'o-')
plt.title('1st Graph')
plt.ylabel('Damped oscillation')

plt.subplot(2, 1, 1)                # nrows=2, ncols=1, index=2
plt.plot(x2, y2, '.-')
plt.title('2nd Graph')
plt.xlabel('time (s)')
plt.ylabel('Undamped')

plt.tight_layout()
plt.show()
```
