```python
import pandas as pd
import matplotlib.pyplot as plt

#       
from matplotlib import font_manager, rc
font_name = font_manager.FontProperties(fname="c:/Windows/Fonts/malgun.ttf").get_name()
rc('font', family=font_name)

#      minus sign   
import matplotlib as mpl
mpl.rcParams['axes.unicode_minus'] = False

# CSV   (  )
df = pd.read_csv('netflix_titles.csv')

# 
mv = df[df['type'] == 'Movie']
#  
mv_cnt = mv.groupby('release_year').size()
#  10
x = mv_cnt.sort_values(ascending=False).head(10)

ratio = x.values  #  
labels = x.index  # 
colors = ['#ff9999', '#ffc000', '#8fd9b6', '#d395d0', '#9dd757', '#d16e79', '#dbda6c', '#2cdeda', '#6c6666', '#799c67']  #  
wedgeprops = {'width': 0.6, 'edgecolor': 'white', 'linewidth': 4}

#    (  )
plt.pie(ratio, labels=[f'{label}: {value}' for label, value in zip(labels, ratio)], colors=colors, wedgeprops=wedgeprops, autopct='%.1f%%' )

#   
plt.title('   TOP10')

#  
plt.show()
```
