```python
import pandas as pd
import matplotlib.pyplot as plt



data = pd.read_csv('netflix_titles.csv')

# 2010년 이전의 데이터를 'Before 2010'으로 그룹화
data['release_year_group'] = data['release_year'].apply(lambda x: 'Before 2010' if x < 2010 else str(x))
release_year_counts = data['release_year_group'].value_counts()
release_year_ratios = release_year_counts / release_year_counts.sum() * 100

plt.bar(release_year_ratios.index, release_year_ratios.values, color='skyblue')
plt.title('연도별 출시된 작품들의 비율')
plt.xlabel('Release Year')
plt.ylabel('Percentage (%)')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import font_manager, rc

# 한글 폰트 설정
font_name = font_manager.FontProperties(fname="c:/Windows/Fonts/malgun.ttf").get_name()
rc('font', family=font_name)

# CSV 파일 읽기
data = pd.read_csv('netflix_titles.csv')

# 연도 범주 정의
bins = [-float('inf'), 2010, 2015, 2020, float('inf')]
labels = ['2010이전에 출시된 영화', '2010년대 전반기 영화', '2010년대 후반기 영화', '2020년대 초반기 영화']

# 'release_year' 컬럼을 범주로 변환
data['release_period'] = pd.cut(data['release_year'], bins=bins, labels=labels)

# 'release_period' 컬럼별로 데이터 빈도 계산
release_period_counts = data['release_period'].value_counts()
release_period_ratios = release_period_counts / release_period_counts.sum() * 100

# 시각화 설정
fig, ax = plt.subplots(1, 2, figsize=(16, 8))

# 파스텔 색상 팔레트
colors = ['#FF9999', '#66B2FF', '#99FF99', '#FFCC99']

# 파이 차트
ax[0].pie(release_period_ratios, labels=release_period_ratios.index, autopct='%0.1f%%', startangle=90, colors=colors)
ax[0].set_title('영화 출시년도별 비율')

# 막대 그래프
ax[1].bar(release_period_ratios.index, release_period_ratios.values, color=colors)
ax[1].set_title('영화 출시년도별 비율 (막대 그래프)')
ax[1].set_xlabel('출시년도 범주')
ax[1].set_ylabel('비율 (%)')
ax[1].grid(axis='y', linestyle='--', alpha=0.7)  # y축에 그리드 추가

plt.show()
```
