```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import font_manager, rc

# 한글 폰트 설정
font_path = 'c:/Windows/Fonts/malgun.ttf'
font_name = font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font_name)
plt.rcParams['axes.unicode_minus'] = False

# 데이터 읽기
netflix_df = pd.read_csv('netflix_titles.csv')

# 2015년 이전의 데이터는 "기타"로 분류
netflix_df['release_year'] = netflix_df['release_year'].apply(lambda x: '기타' if x < 2015 else x)

# 연도별 영화 개수 세기
year_counts = netflix_df['release_year'].value_counts()

# 파스텔 톤 색상 설정
colors = [
    '#FFB3BA',  # 연한 핑크
    '#FFDFBA',  # 연한 오렌지
    '#FFFFBA',  # 연한 노란색
    '#BAFFC9',  # 연한 민트
    '#BAE1FF',  # 연한 하늘색
    '#D5AAFF',  # 연한 보라색
    '#FFDAC1',  # 연한 살구색
    '#C2B6FF',  # 연한 라벤더
    '#B9FBC0',  # 연한 녹색
    '#FFABAB',  # 연한 코랄
    '#B9FBC0',  # 연한 민트
    '#D3B5FF'   # 연한 보라
]

# "기타" 항목을 강조하기 위해 explode 리스트 설정
explode = [0.1 if label == '기타' else 0 for label in year_counts.index]

# 파이차트 생성
plt.figure(figsize=(10, 10))  # 사이즈 설정


plt.pie(
    year_counts,
    labels=year_counts.index,
    colors=colors,
    autopct='%1.1f%%',
    startangle=140,
    explode=explode  # "기타" 항목을 강조
)

# 범례 추가
plt.legend(
    labels=year_counts.index,
    loc='upper right',
    bbox_to_anchor=(1.2, 1),
    title='연도',
    title_fontsize='13',
    fontsize='11'
)

plt.title('연도별 넷플릭스 작품 수')

# 차트 출력
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import font_manager, rc

# 한글 폰트 설정
font_path = 'c:/Windows/Fonts/malgun.ttf'
font_name = font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font_name)
plt.rcParams['axes.unicode_minus'] = False

# 데이터 읽기
netflix_df = pd.read_csv('netflix_titles.csv')

# 국가별 출현 횟수 계산 및 상위 10개 선택
top_countries = netflix_df['country'].value_counts().head(10)

# 색상 목록 정의
colors = plt.cm.tab10.colors  # 기본적인 10색 팔레트 사용

# 서브플롯 생성 및 사이즈 조정
fig, ax = plt.subplots(1, 2, figsize=(16, 8))  # 전체 그림의 크기 조정

# 막대 그래프 그리기
ax[0].bar(top_countries.index, top_countries.values, color=colors[:len(top_countries)])
ax[0].set_xlabel('국가', fontsize=14)
ax[0].set_ylabel('개수', fontsize=14)
ax[0].set_title('상위 10개 국가별 넷플릭스 타이틀 수', fontsize=16)
ax[0].tick_params(axis='x', rotation=45)

# 값 표시
for p in ax[0].patches:
    ax[0].annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2, p.get_height()), 
                   ha='center', va='center', xytext=(0, 5), textcoords='offset points', fontsize=12)

# 파이 차트 그리기
ax[1].pie(top_countries, labels=top_countries.index, autopct='%1.1f%%', colors=colors[:len(top_countries)], startangle=140)

# 자동 레이아웃 조정
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# 데이터 읽기
netflix_df = pd.read_csv('netflix_titles.csv')

# 'rating' 열만 추출
ratings = netflix_df['rating'].dropna()  # NaN 값 제거

# 빈 데이터프레임 생성하여 'rating' 열의 값들을 카운트
rating_counts = ratings.value_counts()

# 내림차순으로 정렬
rating_counts = rating_counts.sort_values(ascending=False)

# 색상을 랜덤하게 생성하거나 지정할 수 있습니다.
colors = plt.cm.viridis(np.linspace(0, 1, len(rating_counts)))

# 가로 막대 그래프 그리기
plt.figure(figsize=(12, 8))
bars = plt.barh(rating_counts.index, rating_counts.values, color=colors)

# 각 막대 위에 값 표기
for bar in bars:
    plt.text(
        bar.get_width() + 0.5,  # x 좌표 (막대 끝 + 약간의 여백)
        bar.get_y() + bar.get_height() / 2,  # y 좌표 (막대의 중앙)
        f'{int(bar.get_width())}',  # 막대의 값
        va='center',  # 수직 정렬
        ha='left',  # 수평 정렬
        color='black'  # 글자 색상
    )

# 제목과 레이블 추가
plt.title('넷플릭스 등급 분포도')
plt.xlabel('개수')
plt.ylabel('등급')

# 그래프 표시
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 데이터 읽기
netflix_df = pd.read_csv('netflix_titles.csv')

# 프로그램 길이 데이터 준비
def extract_duration(duration):
    try:
        if 'min' in duration:
            return int(duration.split()[0])
        elif 'Season' in duration:
            return int(duration.split()[0]) * 30  # Assume 30 minutes per episode as an average
        else:
            return 0
    except:
        return 0

netflix_df['duration_minutes'] = netflix_df['duration'].apply(extract_duration)

# 날짜 데이터를 활용하기 위해 날짜 컬럼을 추가
# 'release_year' 컬럼을 날짜 형식으로 변환
netflix_df['release_year'] = pd.to_datetime(netflix_df['release_year'], format='%Y', errors='coerce')
netflix_df['release_year'] = netflix_df['release_year'].dt.year

# 1990년 이후의 데이터 필터링
filtered_df = netflix_df[netflix_df['release_year'] >= 1990]

# 연도별 평균 프로그램 길이 계산
average_duration_by_year = filtered_df.groupby('release_year')['duration_minutes'].mean()

# 선 그래프 생성
plt.figure(figsize=(12, 8))
plt.plot(average_duration_by_year.index, average_duration_by_year.values, marker='o', linestyle='-', color='b')

plt.xlabel('연도')
plt.ylabel('평균 프로그램 길이 (분)')
plt.title('연도별 평균 프로그램 길이')
plt.grid(True)
plt.xticks(average_duration_by_year.index, rotation=45)  # X축 레이블 회전
plt.tight_layout()  # 레이아웃 조정

# 차트 출력
plt.show()
```

```python
netflix_df = netflix_df[~netflix_df['duration'].str.contains('Season', na=False)]

# 'duration'이 'min'으로 끝나는 문자열인 경우만 처리합니다.
def categorize_duration(duration):
    if 'min' in str(duration):
        try:
            minutes = int(duration.replace(' min', ''))
            return 'long' if minutes >= 100 else 'short'
        except ValueError:
            return None
    return None

# 'Category' 열을 추가합니다.
netflix_df['Category'] = netflix_df['duration'].apply(categorize_duration)
netflix_df.head(20)
```
