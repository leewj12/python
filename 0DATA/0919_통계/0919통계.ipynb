```python
#auto - mpq 

from scipy import stats
from statsmodels.formula.api import ols, glm

auto = pd.read_csv('auto-mpg.csv', sep = ',', header = 0, engine = 'python')

# '?'  NaN  ,  
auto['horsepower'] = pd.to_numeric(auto['horsepower'], errors='coerce')
auto.dropna(inplace=True)

x = auto.drop(columns = ['mpg','car_name'])
# print(x)

#   R- 
Rformula = 'mpg ~ cylinders + displacement + horsepower + weight + acceleration + model_year + origin'

#    (OLS) 
ols_res = ols(Rformula, data = auto).fit()

#     
print(ols_res.summary())

# 'quality' 'type'    
sample1 = auto[auto.columns.drop(['mpg','car_name'])]

#   5 
sample1 = sample1.head(5)

#   (     )
sample1_predict = ols_res.predict(sample1)

#   
print(sample1_predict)
```

```python
import pandas as pd
from scipy import stats
from statsmodels.formula.api import ols

#  
red_df = pd.read_csv('winequality-red.csv', sep=';', header=0, engine='python')
white_df = pd.read_csv('winequality-white.csv', sep=';', header=0, engine='python')

#       'type'  
red_df['type'] = 'red'
white_df['type'] = 'white'

#  
wine = pd.concat([red_df, white_df])

#   quality   
red_wine_quality = wine.loc[wine['type'] == 'red', 'quality']
white_wine_quality = wine.loc[wine['type'] == 'white', 'quality']

# stats.ttest_ind  t     
t_stat, p_value = stats.ttest_ind(red_wine_quality, white_wine_quality, equal_var=False)
print(f"T-statistic: {t_stat}, P-value: {p_value}")

#   R- 
Rformula = 'quality ~ fixed_acidity + volatile_acidity + citric_acid + \
            residual_sugar + chlorides + free_sulfur_dioxide + \
            total_sulfur_dioxide + density + pH + sulphates + alcohol'

#    (OLS) 
regression_result = ols(Rformula, data=wine).fit()

#     
print(regression_result.summary())

# 'quality' 'type'    
sample1 = wine[wine.columns.difference(['quality', 'type'])]

#   5 
sample1 = sample1.head(5)

#   (     )
sample1_predict = regression_result.predict(sample1)

#   
print(sample1_predict)
```
